# Plotting the equation with one line

```python
import matplotlib.pyplot as plt            #importing the module related to plot and label

import numpy as np                         # importing the number python      
```

```python
x_low = 0                  # minmum value for x = 0
x_high = 50                # maximum value for x = 50
number_of_points = 100     # number of lines in y
```

```python
x = np.linspace(x_low,x_high,number_of_points)
y = np.zeros(number_of_points)
```

```python
#print(x)
#print(y)
```

```python
y = 2*x + 3
```

```python
#print(y)
```

```python
plt.plot(x,y, color="red")       # plotting the values and color
plt.xlabel("x axis")             # name of the label for x
plt.ylabel("y axis")             # name of the label for y
plt.title("Line plot")           # name of the plot 
#plt.show()                      # to show figure on the notebook
plt.savefig("myplot.jpg",bbox_inches="tight")    # saving the image with its name and resolution
```

# Plotting two lines in same plot

```python
z = np.zeros(number_of_points)
z = 6*x - 1
```

```python
plt.plot(x,y, color = "red")
plt.plot(x,z, color = "blue")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line plot")
#plt.show()
plt.savefig("myplot 2.jpg")
```
